```python
from scipy.optimize import curve_fit

# Define a custom function to fit
def custom_function(x, a, b, c):
    return a * x**2 + b * x + c

# Fit the custom function to the data
params, covariance = curve_fit(custom_function, x_values, y_values)

# Get the fitted parameters
a, b, c = params

# Generate y values for the trendline
trendline_y = custom_function(x_values, a, b, c)

# Plot the trendline
plt.plot(x_values, trendline_y, label='Custom Trendline', color='purple')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the sequences
sequence1 = np.array([1, 2, 10, 154, 10334, 5399325])
sequence2 = np.array([1, 2, 10, 190, 31730, 240513400])
sequence3 = np.array([1, 2, 10, 130, 5460, 622440])

# Create x values (assuming each sequence starts at index 1)
x_values = np.arange(1, len(sequence1) + 1)

# Create a scatter plot with a logarithmic y-axis
plt.scatter(x_values, sequence1, label=r'$(p_n)_{n \in \mathbb{N}}$', color='blue', marker='o')
plt.scatter(x_values, sequence2, label=r'$\mathcal{PU}$', color='green', marker='s')
plt.scatter(x_values, sequence3, label=r'$\mathcal{PL}$', color='orange', marker='^')

# Add labels and legend
plt.xlabel('n term')
plt.ylabel('Sequence nth term (log scale)')
plt.yscale('log')  # Set y-axis to logarithmic scale
plt.legend()

# Add grid lines for better readability
plt.grid(True, which='both', linestyle='-', linewidth=0.5)

# Customize the appearance
plt.title('Sequence Visualisation')
plt.tight_layout()  # Adjust layout for better spacing

# Show the plot
plt.show()
```
